```python
import time
import matplotlib.pyplot as plt
import math

# -------------------------
# Define Time Complexity Functions
# -------------------------

def constant_time(arr):
    return arr[0]  # O(1)

def linear_time(arr):
    total = 0
    for x in arr:
        total += x
    return total  # O(n)

def quadratic_time(arr):
    total = 0
    for i in arr:
        for j in arr:
            total += i * j
    return total  # O(n^2)

def logarithmic_time(n):
    count = 0
    while n > 1:
        n //= 2
        count += 1
    return count  # O(log n)


# -------------------------
# Input Sizes
# -------------------------

input_sizes = [10, 100, 300, 500]  # Avoid very large n for O(n^2)

times_O1 = []
times_On = []
times_On2 = []
times_Ologn = []

# -------------------------
# Measure Execution Time (in milliseconds)
# -------------------------

for n in input_sizes:
    arr = list(range(n))

    start = time.time()
    constant_time(arr)
    times_O1.append((time.time() - start) * 1000)

    start = time.time()
    linear_time(arr)
    times_On.append((time.time() - start) * 1000)

    start = time.time()
    quadratic_time(arr)
    times_On2.append((time.time() - start) * 1000)

    start = time.time()
    logarithmic_time(n)
    times_Ologn.append((time.time() - start) * 1000)


# -------------------------
# Plot Graph
# -------------------------

plt.figure()

plt.plot(input_sizes, times_O1, label='O(1)')
plt.plot(input_sizes, times_On, label='O(n)')
plt.plot(input_sizes, times_On2, label='O(n^2)')
plt.plot(input_sizes, times_Ologn, label='O(log n)')

plt.yscale('log')  # IMPORTANT: Makes small values visible

plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (milliseconds)")
plt.title("Time Complexity Growth Comparison")
plt.legend()

plt.show()
```

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1


def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1


sizes = [1000, 5000, 10000]
linear_times = []
binary_times = []

for n in sizes:
    arr = list(range(n))

    # Worst case
    start = time.time()
    linear_search(arr, -1)
    linear_times.append(time.time() - start)

    start = time.time()
    binary_search(arr, -1)
    binary_times.append(time.time() - start)
```

```python
plt.plot(sizes, linear_times, label="Linear Search")
plt.plot(sizes, binary_times, label="Binary Search")

plt.xlabel("Array Size")
plt.ylabel("Execution Time")
plt.title("Linear vs Binary Search (Worst Case)")
plt.legend()
plt.show()

```

```python
factorial_calls = 0

def factorial(n):
    global factorial_calls
    factorial_calls += 1

    if n == 0:
        return 1
    return n * factorial(n - 1)

```

```python
fib_recursive_calls = 0

def fib_recursive(n):
    global fib_recursive_calls
    fib_recursive_calls += 1

    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)

```

```python
def fib_dp(n):
    if n <= 1:
        return n

    dp = [0] * (n + 1)
    dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]
```

```python
n_values = [5, 10, 20]

print("n | Factorial Calls | Fact Time | Fib Rec Calls | Fib Rec Time | Fib DP Time")
print("-"*80)

for n in n_values:
    # Factorial
    factorial_calls = 0
    start = time.time()
    factorial(n)
    fact_time = time.time() - start

    # Fibonacci Recursive
    fib_recursive_calls = 0
    start = time.time()
    fib_recursive(n)
    fib_rec_time = time.time() - start

    # Fibonacci DP
    start = time.time()
    fib_dp(n)
    fib_dp_time = time.time() - start

    print(f"{n} | {factorial_calls:^15} | {fact_time:.6f} | "
          f"{fib_recursive_calls:^14} | {fib_rec_time:.6f} | {fib_dp_time:.6f}")
```

```python
import time

calls_rec1 = 0

def recurrence_1(n):
    global calls_rec1
    calls_rec1 += 1

    if n <= 1:
        return n
    return recurrence_1(n // 2) + n
```

```python
calls_rec2 = 0

def recurrence_2(n):
    global calls_rec2
    calls_rec2 += 1

    if n <= 1:
        return n
    return recurrence_2(n // 2) + recurrence_2(n // 2) + n
```

```python
n_values = [10, 100, 500, 1000]

print("n | Rec1 Calls | Rec1 Time | Rec2 Calls | Rec2 Time")
print("-" * 60)

for n in n_values:
    # Recurrence 1
    calls_rec1 = 0
    start = time.time()
    recurrence_1(n)
    time_rec1 = time.time() - start

    # Recurrence 2
    calls_rec2 = 0
    start = time.time()
    recurrence_2(n)
    time_rec2 = time.time() - start

    print(f"{n:<4}| {calls_rec1:^10} | {time_rec1:.6f} | "
          f"{calls_rec2:^10} | {time_rec2:.6f}")
```

```python
# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
    return arr


# Selection Sort
def selection_sort(arr):
    n = len(arr)
    
    for i in range(n):
        min_index = i
        
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        
        arr[i], arr[min_index] = arr[min_index], arr[i]
    
    return arr


# Main Program
arr = [64, 34, 25, 12, 22, 11, 90]

print("Original Array:", arr)

bubble_result = bubble_sort(arr.copy())
print("After Bubble Sort:", bubble_result)

selection_result = selection_sort(arr.copy())
print("After Selection Sort:", selection_result)
```
